```python
# import libraries
import pandas as pd
import geopandas as gpd
import numpy as np
import io
import matplotlib.pyplot as plt
```

```python
# create url variable using github url
url = "https://raw.githubusercontent.com/cameron-manning/up221-cameron/refs/heads/main/data/la_county_voter_rates.csv"

# create dataframe
df = pd.read_csv(url)
```

```python
## initial data exploration
# shape of the dataframe
print(f"The dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")
```

```python
# info about dataframe variables and observations
print("\nDataset Info Summary:")
buffer = io.StringIO()
df.info(buf = buffer)
info_output = buffer.getvalue().splitlines()

# extract column details from df.info()
for line in info_output[5:-2]:  # skipping the first and last lines of df.info() output
    print(line.strip())
```

```python
# first ten rows of the dataset
print(f"The first 10 rows of the dataset are displayed below to give an idea of the structure and content:\n")
print(df.head(10).to_string(index=False))  # to_string() for better alignmentprint("-" * 50)
```

```python
# identify unique years and their counts
year_counts = df["year"].value_counts()
print("Unique years and their counts in the dataset:")
print(year_counts)

# create unique_years from year_counts
unique_years = year_counts.index
print("\nUnique years in the dataset:", list(unique_years))
```

```python
# create scatterplots for each year of data
for year in unique_years:
    # filter the dataframe for the current year
    year_df = df[df["year"] == year]
    
    # create a scatter plot
    plt.figure(figsize=(8, 6)) 
    plt.scatter(year_df["registration_pct"], year_df["voter_turnout_pct"], color="blue", alpha=0.7)
    plt.title(f"Voter Turnout for {year}", fontsize=14)
    plt.xlabel("Voter Registration (%)", fontsize=12)
    plt.ylabel("Voter Turnout (%)", fontsize=12)
    plt.xticks(rotation=45, ha="right", fontsize=10)  # rotate x-axis labels for readability
    plt.grid(axis="y", linestyle="--", alpha=0.7)  # add gridlines
    plt.tight_layout()
    plt.show()
```
